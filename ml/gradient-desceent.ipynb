```python
%matplotlib inline

import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.colors as mat_c


def gradient_descent(x: np.array, y: np.array):
    m_i = b_i = 0
    iterations = 10
    n = len(x)
    learning_rate = 0.08

    plt.scatter(x, y, color='red', marker='+', linewidth=5)

    mse_cost_function = lambda: 1/n*sum([val ** 2 for val in delta_y])
    prev_cost = 0

    for i in range(iterations):
        y_predicted = m_i*x + b_i
        delta_y = (y - y_predicted)
        cost_i = mse_cost_function()
        md = -(2/n)*sum(x*delta_y)
        bd = -(2/n)*sum(delta_y)
        m_i = m_i - learning_rate*md
        b_i = b_i - learning_rate*bd
        print(f'm {m_i}, b {b_i}, cost {cost_i}, iter {i}')
        plt.plot(x, y_predicted, color='green')
        if math.isclose(prev_cost, cost_i, abs_tol=10 ** -30):
            print(f"breaking at iter-{i}")
            break

        prev_cost = cost_i


```

```python
# x,y vectors
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
gradient_descent(x, y)
```
